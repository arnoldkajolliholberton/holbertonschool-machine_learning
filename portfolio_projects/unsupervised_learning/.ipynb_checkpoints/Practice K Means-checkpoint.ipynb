## 3: Clustering with K-Means

Now that we have explored the data, the next logical step is **clustering**â€”grouping customers based on similar characteristics. One of the most popular clustering algorithms is **K-Means**, which we will apply to segment mall customers.  

### **What is K-Means?**  
K-Means is an **unsupervised machine learning algorithm** used to partition data into **K distinct groups (clusters)**. It works by:  
1. **Choosing K cluster centers (centroids)** randomly.  
2. **Assigning each data point to the nearest centroid** based on distance.  
3. **Updating the centroids** by computing the average position of all points in each cluster.  
4. **Repeating steps 2 and 3** until clusters no longer change significantly (convergence).  

### **Why Use K-Means for Mall Customers?**  
- **Uncovers customer segments:**  
  - Groups customers with similar behaviors (e.g., high spenders, budget shoppers).  
- **Helps with targeted marketing:**  
  - Businesses can tailor promotions based on customer types.  
- **Identifies trends and spending patterns:**  
  - Clusters may reveal insights such as **young high spenders** vs. **older conservative shoppers**.  
- **Enhances business strategy:**  
  - Personalized recommendations, loyalty programs, and store layout optimizations can be designed based on cluster insights.  

### **How Will We Apply K-Means Here?**  
We will apply K-Means clustering using customer data, specifically focusing on:  
1. **Choosing the right features** (e.g., Annual Income, Spending Score) to form meaningful clusters.  
2. **Determining the optimal number of clusters (K)** using methods like the **Elbow Method**.  
3. **Applying K-Means and visualizing the clusters** to interpret the results.  

### **Expected Outcome**  
By clustering customers, we aim to create well-defined **customer segments** that businesses can use for better decision-making. These insights will be **data-driven**, allowing for more effective marketing strategies and customer engagement.

## K-Means Clustering: Annual Income & Spending Score

#### **Why These Features?**  
We use **Annual Income (k$) and Spending Score (1-100)** because:  
- They capture **spending behavior** and **income level**, which are key for customer segmentation.  
- They naturally form **well-separated clusters**, making visualization easier.  
- These two features are **not highly correlated** (as seen in the heatmap), ensuring meaningful clustering.  

---

### Determine the Optimal Number of Clusters (Elbow Method)
Before applying K-Means, we need to determine the best **K (number of clusters)**. The **Elbow Method** helps by:  
1. Running K-Means for different values of **K**.  
2. Calculating the **inertia (sum of squared distances from points to their nearest centroid)**.  
3. Identifying the "elbow point" where the inertia stops decreasing significantly, indicating the optimal **K**.  

```python
from sklearn.cluster import KMeans

X = data[['Annual Income (k$)', 'Spending Score (1-100)']].values

# Using the Elbow Method to find optimal K
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(6, 4))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

### Apply K-Means with Optimal K & Visualize Clusters

Once the best `K` is found (in our case `k=5``, we:

- Apply K-Means using that value of K.
- Assign each customer to a cluster.
- Visualize the clusters in a scatter plot.

```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Remove special characters and spaces from column names for easier handling
data.columns = data.columns.str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# Selecting features for clustering
X = data[['Annual_Income_k$', 'Spending_Score_1-100']]

# Applying K-Means with optimal k
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
data['Cluster'] = kmeans.fit_predict(X)

# Scatter plot with bigger figure size
plt.figure(figsize=(10, 8))  # Increase figure size

sns.scatterplot(x='Annual_Income_k$', 
                y='Spending_Score_1-100', 
                hue='Cluster',  # Using cluster numbers as hue
                data=data,      # Pass the data to sns.scatterplot
                palette='Set1', 
                s=100, 
                alpha=0.8)

# Plot cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', marker='X', s=200, label='Centroids')

plt.title('K-Means Clustering: Annual Income vs Spending Score', fontsize=16)
plt.xlabel('Annual Income (k$)', fontsize=14)
plt.ylabel('Spending Score (1-100)', fontsize=14)
plt.legend(title='Clusters', fontsize=12)
plt.show()
```

### **Cluster Validation with Silhouette Score and Agglomerative Clustering**

Cluster validation is essential to assess the quality of the clusters you've created. The **Silhouette Score** can help measure the cohesion and separation of clusters, while **Agglomerative Clustering** can be used as an alternative clustering method to compare results.

#### **1. Silhouette Score**
The **Silhouette Score** is a metric used to determine how similar an object is to its own cluster compared to other clusters. It ranges from -1 to 1:
- **1**: Perfect clustering (data points are very similar within their own cluster and dissimilar to other clusters).
- **0**: Indifferent (data points are on or near the decision boundary between clusters).
- **-1**: Poor clustering (data points are misclassified).

**Formula:**
The silhouette score for each point \( i \) is calculated as:
$$
S(i) = \frac{b(i) - a(i)}{ \max(a(i), b(i)) }
$$
Where:
- $a(i)$ is the average distance between $i$ and all other points in the same cluster.
- $b(i)$ is the average distance between $i$ and all points in the nearest cluster.

A **higher Silhouette Score** indicates better-defined and well-separated clusters.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
data['Cluster_KMeans'] = kmeans.fit_predict(X)

silhouette_kmeans = silhouette_score(X, data['Cluster_KMeans'])
print(f'Silhouette Score for K-Means: {silhouette_kmeans}')
```

The Silhouette Score of 0.55 for K-Means indicates that the clusters are moderately well-separated. A score closer to 1 would suggest very distinct clusters, while a score near 0 indicates overlapping clusters. In this case, the score of 0.55 shows that the clustering is decent, but there is still room for improvement in terms of separation between clusters.

### Agglomerative Clustering:

Agglomerative Clustering is a type of hierarchical clustering algorithm. Unlike K-Means, which requires the number of clusters to be specified upfront, Agglomerative Clustering builds a hierarchy of clusters by iteratively merging the closest pair of clusters. Itâ€™s particularly useful when you donâ€™t know how many clusters you should expect and is better suited for non-spherical cluster shapes. The result is often represented as a dendrogram, which shows the hierarchy of clusters.

Agglomerative Clustering is valuable when you need a hierarchical structure of your data or when you want to explore potential cluster relationships without predefining the number of clusters. It can be used for discovering patterns that K-Means might miss due to its rigid structure of specifying cluster numbers.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage

# Clean column names
data.columns = data.columns.str.replace(' ', '_').str.replace('($)', '').str.replace('[^A-Za-z0-9_]', '', regex=True)

# Select relevant features for clustering
X = data[['Annual_Income_k', 'Spending_Score_1100']]

# Standardize the data before clustering (important for distance-based algorithms)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform hierarchical/agglomerative clustering using 'ward' method
Z = linkage(X_scaled, method='ward')

# Plot the dendrogram with different colors for clusters, and hide x-axis labels
plt.figure(figsize=(12, 8))
dendrogram(Z, labels=data['CustomerID'].values, orientation='top', color_threshold=5)

# Hide x-axis labels
plt.xticks([])

plt.title('Dendrogram for Agglomerative Clustering')
plt.xlabel('Customer ID')
plt.ylabel('Distance')
plt.show()
```
